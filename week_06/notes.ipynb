```python
from statistics import mean, median, quantiles
from math import ceil

import matplotlib.pyplot as plt
import pandas as pd
```

### Exercise 01

```python
nmbrs = [2159, 1578, 773, 1681, 462, 3934, 2184, 203, 7395, 58, 513, 1230, 695, 321, 726]
nmbrs.sort()

min = nmbrs[0]
max = nmbrs[-1]

# calculate the mean
mean = sum(nmbrs) / len(nmbrs)

# calculate the median
if len(nmbrs) % 2 == 0:
  median = (nmbrs[len(nmbrs) // 2 - 1] + nmbrs[len(nmbrs) // 2]) / 2
else:
  median = nmbrs[len(nmbrs) // 2]

# calculate 1st and 3rd quartile
q1 = nmbrs[ceil(len(nmbrs) / 4)]
q3 = nmbrs[ceil(len(nmbrs) / 4 * 3)]

print(f"Min: {min}\nQ1: {q1}\nMedian: {median}\nQ3: {q3}\nMax: {max}")
```

```python
# draw a modified boxplot
plot_min = q1 - 1.5 * (q3 - q1)
plot_max = q3 + 1.5 * (q3 - q1)

# draw a boxplot
plt.boxplot(nmbrs, whis=1.5, showmeans=True)
plt.show()
```
